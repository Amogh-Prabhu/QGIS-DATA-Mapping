```python
import pandas as pd
import numpy as np
```

```python
climate_df = pd.read_csv('Copy of datafile.csv')
climate_df
```

```python
new_df = pd.DataFrame()
yearly_min = climate_df.groupby('Station Name')[
    ['Mean Temperature  in degree C - Minimum']]
yearly_max = climate_df.groupby('Station Name')[
    ['Mean Temperature in degree C - Maximum']]
```

```python
new_df['Yearly Average Min Temp'] = round(yearly_min.sum() / 12, 4)
new_df['Yearly Average Max Temp'] = round(yearly_max.sum() / 12, 4)
new_df['Average Rainfall'] = round(climate_df.groupby(
    'Station Name')[['Mean Rainfall in mm']].sum() / 12, 4)

new_df
```

```python
new_df['Range of Min Temp'] = yearly_min.min().astype(str) + " - " + \
    yearly_min.max().astype(str)
new_df['Range of Max Temp'] = yearly_max.min().astype(str) + " - " + \
    yearly_max.max().astype(str)
new_df['Latitude'] = round(climate_df.groupby(
    'Station Name')[['Latitude']].sum() / 12, 4)
new_df['Longitude'] = round(climate_df.groupby(
    'Station Name')[['Longitude']].sum() / 12, 4)

new_df
```

```python
new_df.to_csv('new_datafram.csv')
```
